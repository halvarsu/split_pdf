```python
# Command to generate results.txt: 
# $ pdftk helgaker.pdf dump_data output results.txt
```

```python
import pandas as pd 

with open("results.txt") as infile:
    txt = infile.read()

txt = "\n".join(line for line in txt.split("\n") if "Bookmark" in line)
bms = txt.split("BookmarkBegin")

lines = [dict(v.split(":", 1) for v in bm.split("\n") if v) for bm in bms if bm]
data = []
for bm in lines:
    data.append(pd.Series(bm))
```

```python
df = pd.DataFrame(data)
df['BookmarkLevel'] = pd.to_numeric(df['BookmarkLevel'])
df['BookmarkPageNumber'] = pd.to_numeric(df['BookmarkPageNumber'])
df.columns = ['title', 'level', 'page']
df['snake_title'] = df.title.str.strip().str.replace(" ", "_").str.lower()
df
```

```python
chapter_info = df[df['level'] == 1][['title', 'page']].set_index("page")
chapter_txt = "offset 0\n" + chapter_info.to_string(index_names=False, header=False)
print(chapter_txt)
```

```python
with open("helgaker_chapters.txt", "w") as outfile:
    outfile.write(chapter_txt)
```
